```python
# Univariate analysis

def analyze_target(data,target):

    print(data[target].describe())
    sns.histplot(data[target],kde=True,color='blue',bins=30)
    plt.show()
```

```python
def analyze_num(data,num_col):
    for col in num_col:
        plt.figure(figsize=(5,5))
        sns.boxplot(data=data[col])
        plt.title(f'Boxplot of {col}')
        plt.show()

analyze_num(df1,num)

# sns.boxplot(data=df1['SalePrice'])
# plt.show()
```

```python
def analyze_cat(data,category):
    for cats in category:
        plt.Figure(figsize=(5,5))
        sns.countplot(data=data[cats],order=data[cats].value_counts().index,palette='viridis')
        plt.title(f'CountPlot of {cats}')
        plt.xticks(rotation=45)
        plt.show()


analyze_cat(df1,cat[:5])
```
