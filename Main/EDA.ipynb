```python
# Hanadeling missing data from numerical columns

df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())
df['MasVnrArea'] = df['MasVnrArea'].fillna(df['MasVnrArea'].median())
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())
```

```python
# Handeling missing data from categorical columns

df['GarageQual'] = df['GarageQual'].fillna(df['GarageQual'].mode()[0])
df['GarageFinish'] = df['GarageFinish'].fillna(df['GarageFinish'].mode()[0])
df['GarageType'] = df['GarageType'].fillna(df['GarageType'].mode()[0])
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
df['BsmtFinType2'] = df['BsmtFinType2'].fillna(df['BsmtFinType2'].mode()[0])
df['BsmtFinType1'] = df['BsmtFinType1'].fillna(df['BsmtFinType1'].mode()[0])
df['BsmtExposure'] = df['BsmtExposure'].fillna(df['BsmtExposure'].mode()[0])
df['BsmtCond'] = df['BsmtCond'].fillna(df['BsmtCond'].mode()[0])
df['BsmtQual'] = df['BsmtQual'].fillna(df['BsmtQual'].mode()[0])
df['GarageCond'] = df['GarageCond'].fillna(df['GarageCond'].mode()[0])
```

```python
# Sale Price trend

sns.histplot(df['SalePrice'],kde=True,color='blue',bins=20)
plt.show()
```

```python
numeric_data = df.select_dtypes(include=['float64', 'int64'])
numeric_data
```

```python
correlation_matrix = numeric_data.corr()
```

```python
strong_correlation_features = correlation_matrix['SalePrice'][abs(correlation_matrix['SalePrice']) > 0.3].index
filtered_correlation_matrix = numeric_data[strong_correlation_features].corr()

```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(filtered_correlation_matrix, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)
plt.title('Heatmap of Correlation Matrix for Numeric Features', fontsize=16)
plt.show()
```

```python
# Identifying categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

for col in categorical_columns:
    if df[col].nunique() <= 20:  # Focus on categorical columns with fewer unique values for readability
        print(f"\nColumn: {col}")
        print(df.groupby(col)['SalePrice'].mean().sort_values(ascending=False))
```

```python
# converting categirical columns to numerical columns
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
# encoding categorical values
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

df
```
